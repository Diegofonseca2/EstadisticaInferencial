```python
numerical_cols = ['Edad (años)', 'Estatura (CM)', 'Cantidad de personas']
df_cleaned = df.copy() # Create a copy to avoid modifying the original DataFrame

for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]

    print(f"Outliers for column '{col}':")
    print(outliers[[col]].describe())

    # Remove outliers
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

print("\nDataFrame after removing outliers:")
print(df_cleaned.info())
print("\nSummary statistics after removing outliers:")
print(df_cleaned[numerical_cols].describe())

# Visualize the distributions after removing outliers
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df_cleaned, x=col, kde=True)
    plt.title(f'Distribución de {col} (sin atípicos)')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df_cleaned, y=col)
    plt.title(f'Diagrama de Caja de {col} (sin atípicos)')
    plt.ylabel(col)
    plt.show()
```

```python
from scipy.stats import shapiro

numerical_cols = ['Edad (años)', 'Estatura (CM)', 'Cantidad de personas']

print("Resultados del test de Shapiro-Wilk en df_cleaned:")

for col in numerical_cols:
    # Shapiro-Wilk test is sensitive to sample size. For very large datasets,
    # it might always reject normality. Also, it can be computationally intensive.
    # We'll use a sample for demonstration if the dataset is too large,
    # but for now, let's try on the full cleaned data.
    # If it takes too long or gives memory errors, we can consider sampling.

    # Handle potential issues with constant values after outlier removal
    if df_cleaned[col].nunique() <= 1:
        print(f"Columna '{col}': No se puede realizar el test de Shapiro-Wilk en datos constantes o con un solo valor único.")
        continue

    try:
        stat, p = shapiro(df_cleaned[col])
        print(f"Columna '{col}': Estadístico = {stat:.4f}, p-valor = {p:.4f}")

        # Interpret the result
        alpha = 0.05
        if p > alpha:
            print(f"La columna '{col}' parece seguir una distribución normal (no se rechaza H0)\n")
        else:
            print(f"La columna '{col}' no parece seguir una distribución normal (se rechaza H0)\n")
    except Exception as e:
        print(f"No se pudo realizar el test de Shapiro-Wilk para la columna '{col}': {e}\n")
```

```python
from scipy.stats import kstest, norm

numerical_cols = ['Edad (años)', 'Estatura (CM)', 'Cantidad de personas']

print("Resultados del test de Kolmogorov-Smirnov en df_cleaned:")

for col in numerical_cols:
    # Kolmogorov-Smirnov test requires data with more than one unique value.
    if df_cleaned[col].nunique() <= 1:
        print(f"Columna '{col}': No se puede realizar el test de Kolmogorov-Smirnov en datos constantes o con un solo valor único.")
        continue

    try:
        # Perform the Kolmogorov-Smirnov test against a normal distribution
        # We need to specify the parameters (mean and standard deviation) of the normal distribution
        # for comparison. We'll use the mean and std dev of the data itself.
        mean = df_cleaned[col].mean()
        std = df_cleaned[col].std()

        # Handle the case where standard deviation is zero (e.g., all values are the same)
        if std == 0:
             print(f"Columna '{col}': La desviación estándar es cero, no se puede realizar el test de Kolmogorov-Smirnov contra una distribución normal.")
             continue


        stat, p = kstest(df_cleaned[col], 'norm', args=(mean, std))
        print(f"Columna '{col}': Estadístico = {stat:.4f}, p-valor = {p:.4f}")

        # Interpret the result
        alpha = 0.05
        if p > alpha:
            print(f"La columna '{col}' parece seguir una distribución normal (no se rechaza H0)\n")
        else:
            print(f"La columna '{col}' no parece seguir una distribución normal (se rechaza H0)\n")
    except Exception as e:
        print(f"No se pudo realizar el test de Kolmogorov-Smirnov para la columna '{col}': {e}\n")
```

```python
from scipy.stats import anderson

numerical_cols = ['Edad (años)', 'Estatura (CM)', 'Cantidad de personas']

print("Resultados del test de Anderson-Darling en df_cleaned:")

for col in numerical_cols:
    # Anderson-Darling test requires data with more than one unique value.
    if df_cleaned[col].nunique() <= 1:
        print(f"Columna '{col}': No se puede realizar el test de Anderson-Darling en datos constantes o con un solo valor único.")
        continue

    try:
        result = anderson(df_cleaned[col], dist='norm')
        print(f"Columna '{col}':")
        print(f"Estadístico = {result.statistic:.4f}")
        print("Valores críticos:")
        for i in range(len(result.critical_values)):
            sl, cv = result.significance_level[i], result.critical_values[i]
            print(f"  {sl}%: {cv:.4f} (La distribución es normal si el estadístico <= este valor)")
        print("\n")

    except Exception as e:
        print(f"No se pudo realizar el test de Anderson-Darling para la columna '{col}': {e}\n")
```

```python
from scipy.stats import jarque_bera

numerical_cols = ['Edad (años)', 'Estatura (CM)', 'Cantidad de personas']

print("Resultados del test de Jarque-Bera en df_cleaned:")

for col in numerical_cols:
    # Jarque-Bera test requires data with more than one unique value and sufficient non-zero variance.
    if df_cleaned[col].nunique() <= 1:
        print(f"Columna '{col}': No se puede realizar el test de Jarque-Bera en datos constantes o con un solo valor único.")
        continue

    try:
        stat, p = jarque_bera(df_cleaned[col])
        print(f"Columna '{col}': Estadístico = {stat:.4f}, p-valor = {p:.4f}")

        # Interpret the result
        alpha = 0.05
        if p > alpha:
            print(f"La columna '{col}' parece seguir una distribución normal (no se rechaza H0)\n")
        else:
            print(f"La columna '{col}' no parece seguir una distribución normal (se rechaza H0)\n")
    except Exception as e:
        print(f"No se pudo realizar el test de Jarque-Bera para la columna '{col}': {e}\n")
```
