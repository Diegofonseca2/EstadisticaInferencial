```python
import scipy.stats as st
import math

# Datos proporcionados (con estudio piloto)
proporcion_piloto = 0.72  # Proporción del estudio piloto
margen_error = 0.03  # Margen de error deseado
nivel_confianza = 0.95  # Nivel de confianza

# Calcula el valor crítico Z para un nivel de confianza del 95%
# Para un intervalo de confianza del 95%, el alfa (α) es 1 - 0.95 = 0.05.
# El valor crítico Z se busca para α/2, que es 0.025.
valor_z = st.norm.ppf(1 - (1 - nivel_confianza) / 2)

# Calcula el tamaño de muestra con estudio piloto
# Fórmula: n = (Z^2 * p * (1-p)) / E^2
tamano_muestra_piloto = (valor_z**2 * proporcion_piloto * (1 - proporcion_piloto)) / (margen_error**2)

# Redondear hacia arriba para asegurar el margen de error deseado
tamano_muestra_piloto = math.ceil(tamano_muestra_piloto)

print("Con estudio piloto:")
print(f"Proporción del estudio piloto: {proporcion_piloto}")
print(f"Margen de error deseado: {margen_error}")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Valor crítico Z: {valor_z:.2f}")
print(f"Tamaño de muestra necesario: {tamano_muestra_piloto}")

print("\n" + "="*30 + "\n")

# Datos proporcionados (sin estudio piloto - enfoque conservador)
proporcion_conservadora = 0.5  # Proporción conservadora (p=0.5 maximiza n)
# Margen de error y nivel de confianza son los mismos

# Calcula el tamaño de muestra sin estudio piloto
# Fórmula: n = (Z^2 * p * (1-p)) / E^2
tamano_muestra_conservadora = (valor_z**2 * proporcion_conservadora * (1 - proporcion_conservadora)) / (margen_error**2)

# Redondear hacia arriba
tamano_muestra_conservadora = math.ceil(tamano_muestra_conservadora)

print("Sin estudio piloto (enfoque conservador):")
print(f"Proporción conservadora (p=0.5): {proporcion_conservadora}")
print(f"Margen de error deseado: {margen_error}")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Valor crítico Z: {valor_z:.2f}")
print(f"Tamaño de muestra necesario: {tamano_muestra_conservadora}")
```

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 500 (El peso promedio es 500 g)
# H1: μ < 500 (El peso promedio es menor a 500 g - subllenado)

# Datos proporcionados
media_muestral = 495  # Media muestral
media_poblacional_bajo_H0 = 500  # Media poblacional bajo la hipótesis nula
desviacion_estandar_poblacional = 15  # Desviación estándar poblacional
tamano_muestra = 40  # Tamaño de la muestra
nivel_significancia = 0.05  # Nivel de significancia (α)

# 2. Calcular el estadístico Z
# Fórmula: Z = (x̄ - μ) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / (tamano_muestra**0.5)
estadistico_z = (media_muestral - media_poblacional_bajo_H0) / error_estandar

# 3. Determinar la región crítica y la región de rechazo
# Para una prueba unilateral izquierda (H1: μ < 500), el valor crítico Z se busca para α.
valor_critico_z = st.norm.ppf(nivel_significancia)

# 4. Concluir si se rechaza H0
print(f"Hipótesis Nula (H0): μ = {media_poblacional_bajo_H0} g")
print(f"Hipótesis Alternativa (H1): μ < {media_poblacional_bajo_H0} g")
print(f"Media muestral: {media_muestral} g")
print(f"Desviación estándar poblacional: {desviacion_estandar_poblacional} g")
print(f"Tamaño de la muestra: {tamano_muestra}")
print(f"Nivel de significancia (α): {nivel_significancia}")

print(f"\nEstadístico Z calculado: {estadistico_z:.2f}")
print(f"Valor crítico Z para α = {nivel_significancia}: {valor_critico_z:.2f}")

# Conclusión
if estadistico_z < valor_critico_z:
    conclusion = "Se rechaza la hipótesis nula (H0)."
    implicacion = "Hay evidencia significativa para concluir que el peso promedio de los paquetes es menor a 500 g."
else:
    conclusion = "No se rechaza la hipótesis nula (H0)."
    implicacion = "No hay suficiente evidencia para concluir que el peso promedio de los paquetes es menor a 500 g."

print(f"\nConclusión: {conclusion}")
print(f"Implicación: {implicacion}")

# Gráfica de la distribución normal y la región de rechazo
x = np.linspace(-4, 4, 1000)
plt.plot(x, st.norm.pdf(x, 0, 1), label='Distribución Normal Estándar')
plt.fill_between(x, 0, st.norm.pdf(x, 0, 1), where=(x < valor_critico_z), color='red', alpha=0.5, label='Región de Rechazo (α)')
plt.axvline(estadistico_z, color='blue', linestyle='dashed', linewidth=2, label=f'Estadístico Z ({estadistico_z:.2f})')
plt.axvline(valor_critico_z, color='red', linestyle='dashed', linewidth=2, label=f'Valor Crítico Z ({valor_critico_z:.2f})')
plt.title('Test de Hipótesis Unilateral Izquierda')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: σ^2 = 2.5 (La varianza poblacional es 2.5 minutos^2)
# H1: σ^2 ≠ 2.5 (La varianza poblacional difiere de 2.5 minutos^2) - prueba bilateral

# Datos proporcionados
varianza_poblacional_bajo_H0 = 2.5  # Varianza poblacional bajo la hipótesis nula
varianza_muestral = 4  # Varianza muestral
tamano_muestra = 16  # Tamaño de la muestra
nivel_significancia = 0.05  # Nivel de significancia (α)

# Grados de libertad
grados_libertad = tamano_muestra - 1

# 2. Calcule el estadístico χ²
# Fórmula: χ² = ((n - 1) * s^2) / σ^2
estadistico_chi2 = (grados_libertad * varianza_muestral) / varianza_poblacional_bajo_H0

# 3. Determine los grados de libertad y los valores críticos
# Para una prueba bilateral (H1: σ² ≠ 2.5), el alfa (α) se divide por 2.
# Los valores críticos Chi-cuadrado se buscan para α/2 y 1 - α/2, y los grados de libertad.
valor_critico_chi2_inferior = st.chi2.ppf(nivel_significancia / 2, grados_libertad)
valor_critico_chi2_superior = st.chi2.ppf(1 - nivel_significancia / 2, grados_libertad)

# 4. Concluya si se rechaza H0
print(f"Hipótesis Nula (H0): σ² = {varianza_poblacional_bajo_H0} minutos²")
print(f"Hipótesis Alternativa (H1): σ² ≠ {varianza_poblacional_bajo_H0} minutos²")
print(f"Varianza muestral (s²): {varianza_muestral} minutos²")
print(f"Tamaño de la muestra (n): {tamano_muestra}")
print(f"Nivel de significancia (α): {nivel_significancia}")
print(f"Grados de libertad: {grados_libertad}")

print(f"\nEstadístico χ² calculado: {estadistico_chi2:.2f}")
print(f"Valores críticos χ² para α = {nivel_significancia} y {grados_libertad} GL: [{valor_critico_chi2_inferior:.2f}, {valor_critico_chi2_superior:.2f}]")

# Conclusión
if estadistico_chi2 < valor_critico_chi2_inferior or estadistico_chi2 > valor_critico_chi2_superior:
    conclusion = "Se rechaza la hipótesis nula (H0)."
    implicacion = "Hay evidencia significativa para concluir que la varianza poblacional del tiempo de duración de las llamadas difiere de 2.5 minutos²."
else:
    conclusion = "No se rechaza la hipótesis nula (H0)."
    implicacion = "No hay suficiente evidencia para concluir que la varianza poblacional del tiempo de duración de las llamadas difiere de 2.5 minutos²."

print(f"\nConclusión: {conclusion}")
print(f"Implicación: {implicacion}")

# Gráfica de la distribución Chi-cuadrado y la región de rechazo
x = np.linspace(0, st.chi2.ppf(0.999, grados_libertad) + 5, 1000) # Ajustar el rango de x
plt.plot(x, st.chi2.pdf(x, grados_libertad), label=f'Distribución χ² ({grados_libertad} GL)')
plt.fill_between(x, 0, st.chi2.pdf(x, grados_libertad), where=(x < valor_critico_chi2_inferior), color='red', alpha=0.5, label='Región de Rechazo (α/2)')
plt.fill_between(x, 0, st.chi2.pdf(x, grados_libertad), where=(x > valor_critico_chi2_superior), color='red', alpha=0.5, label='Región de Rechazo (α/2)')
plt.axvline(estadistico_chi2, color='blue', linestyle='dashed', linewidth=2, label=f'Estadístico χ² ({estadistico_chi2:.2f})')
plt.axvline(valor_critico_chi2_inferior, color='red', linestyle='dashed', linewidth=2, label=f'Valor Crítico χ² Inferior ({valor_critico_chi2_inferior:.2f})')
plt.axvline(valor_critico_chi2_superior, color='red', linestyle='dashed', linewidth=2, label=f'Valor Crítico χ² Superior ({valor_critico_chi2_superior:.2f})')
plt.title('Test de Hipótesis Bilateral (Varianza Tiempo Llamadas)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: σ^2 = 1.2 (La varianza poblacional es 1.2 ml^2)
# H1: σ^2 > 1.2 (La varianza poblacional es mayor a 1.2 ml^2) - prueba unilateral derecha

# Datos proporcionados
varianza_poblacional_bajo_H0 = 1.2  # Varianza poblacional bajo la hipótesis nula
varianza_muestral = 2.0  # Varianza muestral
tamano_muestra = 20  # Tamaño de la muestra
nivel_significancia = 0.01  # Nivel de significancia (α)

# Grados de libertad
grados_libertad = tamano_muestra - 1

# 2. Calcule el estadístico χ²
# Fórmula: χ² = ((n - 1) * s^2) / σ^2
estadistico_chi2 = (grados_libertad * varianza_muestral) / varianza_poblacional_bajo_H0

# 3. Determine los grados de libertad y el valor crítico
# Para una prueba unilateral derecha (H1: σ² > 1.2), el valor crítico Chi-cuadrado se busca para 1 - α y los grados de libertad.
valor_critico_chi2 = st.chi2.ppf(1 - nivel_significancia, grados_libertad)

# 4. Concluya si se rechaza H0
print(f"Hipótesis Nula (H0): σ² = {varianza_poblacional_bajo_H0} ml²")
print(f"Hipótesis Alternativa (H1): σ² > {varianza_poblacional_bajo_H0} ml²")
print(f"Varianza muestral (s²): {varianza_muestral} ml²")
print(f"Tamaño de la muestra (n): {tamano_muestra}")
print(f"Nivel de significancia (α): {nivel_significancia}")
print(f"Grados de libertad: {grados_libertad}")

print(f"\nEstadístico χ² calculado: {estadistico_chi2:.2f}")
print(f"Valor crítico χ² para α = {nivel_significancia} y {grados_libertad} GL: {valor_critico_chi2:.2f}")

# Conclusión
if estadistico_chi2 > valor_critico_chi2:
    conclusion = "Se rechaza la hipótesis nula (H0)."
    implicacion = "Hay evidencia significativa para concluir que la varianza poblacional del llenado de botellas es mayor a 1.2 ml²."
else:
    conclusion = "No se rechaza la hipótesis nula (H0)."
    implicacion = "No hay suficiente evidencia para concluir que la varianza poblacional del llenado de botellas es mayor a 1.2 ml²."

print(f"\nConclusión: {conclusion}")
print(f"Implicación: {implicacion}")

# Gráfica de la distribución Chi-cuadrado y la región de rechazo
x = np.linspace(0, st.chi2.ppf(0.999, grados_libertad) + 5, 1000) # Ajustar el rango de x
plt.plot(x, st.chi2.pdf(x, grados_libertad), label=f'Distribución χ² ({grados_libertad} GL)')
plt.fill_between(x, 0, st.chi2.pdf(x, grados_libertad), where=(x > valor_critico_chi2), color='red', alpha=0.5, label='Región de Rechazo (α)')
plt.axvline(estadistico_chi2, color='blue', linestyle='dashed', linewidth=2, label=f'Estadístico χ² ({estadistico_chi2:.2f})')
plt.axvline(valor_critico_chi2, color='red', linestyle='dashed', linewidth=2, label=f'Valor Crítico χ² ({valor_critico_chi2:.2f})')
plt.title('Test de Hipótesis Unilateral Derecha (Varianza Llenado Botellas)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
